```python
from functions import calc_variance
import torch as th
import numpy as np
import matplotlib.pyplot as plt
```

### Slow vs Fast Entangling

```python
ps = np.linspace(0.09, 0.5, 8)
variances1 = [calc_variance(6, 8, p, n_sim=100, fast_ent=True)[1] for p in ps]
variances2 = [calc_variance(6, 20, p, n_sim=100, fast_ent=False)[1] for p in ps]
```

```python
plt.plot(ps, variances1, marker='o', label='fast_ent')
plt.plot(ps, variances2, marker='x', label='slow_ent')
plt.xlabel('p')
plt.ylabel('Variance')
plt.title('Variance vs p')
plt.legend()
plt.show()
```

```python
L = np.arange(1, 11, 1)
variances1 = [calc_variance(5, l, 0.1, n_sim=200, fast_ent=True)[1] for l in L]
variances2 = [calc_variance(5, l, 0.1, n_sim=200, fast_ent=False)[1] for l in L]
```

```python
plt.plot(L, variances1, marker='o', label='fast_ent')
plt.plot(L, variances2, marker='x', label='slow_ent')
plt.xlabel('L')
plt.ylabel('Variance')
plt.yscale('log')
plt.title('Variance vs L')
plt.legend()
plt.show()
```
