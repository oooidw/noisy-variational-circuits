### Slow vs Fast Entangling

```python
layers1 = np.arange(1, 10)
layers2 = np.arange(1, 20, 2)
variances1 = [calc_variance(4, l, n_sim=10, n_sim_noise=30, fast_ent=True)[1] for l in layers1]
# variances2 = [calc_variance(6, l, n_sim=10, n_sim_noise=30, fast_ent=False)[1] for l in layers2]
```

```python
plt.plot(layers1, variances1, marker='o', label='fast_ent')
# plt.plot(layers2, variances2, marker='x', label='slow_ent')
plt.xlabel('Layers')
plt.ylabel('Variance')
plt.title('Variance vs l')
plt.legend()
plt.show()
```

```python
variances1 = [calc_variance(4, l, n_sim=100, n_sim_noise=5, fast_ent=True)[1] for l in layers1]

plt.plot(layers1, variances1, marker='o', label='fast_ent')
plt.xlabel('Layers')
plt.ylabel('Variance')
plt.title('Variance vs l')
plt.legend()
plt.show()
```
