```python
from functions_ngates import calc_variance, layer
import torch as th
import numpy as np
import matplotlib.pyplot as plt
from icecream import ic
```

### Slow vs Fast Entangling

```python
layers1 = np.arange(1, 10)
layers2 = np.arange(1, 10)
variances1 = [calc_variance(4, l, n_sim=500, n_sim_noise=100, fast_ent=True, noise='dephasing')[1] for l in layers1]
variances2 = [calc_variance(4, l, n_sim=500, n_sim_noise=100, fast_ent=True, noise='bitflip')[1] for l in layers1]
variances3 = [calc_variance(4, l, n_sim=500, n_sim_noise=100, fast_ent=True, noise='amplitude_damping')[1] for l in layers1]
```

```python
plt.plot(layers1, variances1, marker='o', label=r'$\sigma_z$')
plt.plot(layers1, variances2, marker='x', label=r'$\sigma_x$')
plt.plot(layers1, variances3, marker='^', label=r'$\sigma_-$')
plt.xlabel('Layers')
plt.ylabel('Variance')
plt.title('Variance vs Layers (Fast)')
plt.legend()
plt.show()
```
